# Task 3.1

Investigate how surface temperature, luminosity, central density and central temperature change with time. How do they change in relation to each other? What does this tell you about what is happening to the star? Include plots to illustrate your answers.

### Surface Temp. vs Time

```python
# plt.figure(figsize=(8,6))
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].plot(data_1_5['Age (years)']/1e9, data_1_5['Log10 Surface temperature (K)'], linestyle='--', c='b')
ax[0].grid();

ax[1].plot(data_20['Age (years)']/1e7, data_20['Log10 Surface temperature (K)'], linestyle='--', c='r')

ax[0].set_xlabel('Age [$10^{9}$ years]')
ax[0].set_ylabel('Log10 Surface temperature [K]')

ax[1].set_xlabel('Age [$10^{7}$ years]')
ax[1].set_ylabel('Log10 Surface temperature [K]')
ax[1].grid();
```

## Luminosity vs Time

### Luminosity from PP Chain

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].plot(data_1_5['Age (years)']/1e9, data_1_5['Luminosity from PP chain (L0)'], linestyle='--', c='b')
ax[0].grid();
ax[1].plot(data_20['Age (years)']/1e7, data_20['Luminosity from PP chain (L0)'], linestyle='--', c='r')

ax[0].set_xlabel('Age [$10^{9}$ years]')
ax[0].set_ylabel('Luminosity from PP chain [L☉]')

ax[1].set_xlabel('Age [$10^{7}$ years]')
ax[1].set_ylabel('Luminosity from PP chain [L☉]')
ax[1].grid();
```

### Luminosity from CNO Cycle

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].plot(data_1_5['Age (years)']/1e9, data_1_5['Luminosity from CNO cycle (L0)']/1e4, linestyle='--', c='b')
ax[0].grid();
ax[1].plot(data_20['Age (years)']/1e7, data_20['Luminosity from CNO cycle (L0)']/1e5, linestyle='--', c='r')

ax[0].set_xlabel('Age [$10^{9}$ years]')
ax[0].set_ylabel('Luminosity from CNO cycle [$10^{4}$ L☉]')

ax[1].set_xlabel('Age [$10^{7}$ years]')
ax[1].set_ylabel('Luminosity from CNO cycle [$10^{5}$ L☉]')
ax[1].grid();
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].plot(data_1_5['Age (years)']/1e9, np.log(data_1_5['Luminosity from CNO cycle (L0)']), linestyle='--', c='b')
ax[0].grid();
ax[1].plot(data_20['Age (years)']/1e7, np.log(data_20['Luminosity from CNO cycle (L0)']), linestyle='--', c='r')

ax[0].set_xlabel('Age [$10^{9}$ years]')
ax[0].set_ylabel('Log Luminosity from CNO cycle [L☉]')

ax[1].set_xlabel('Age [$10^{7}$ years]')
ax[1].set_ylabel('Log Luminosity from CNO cycle [L☉]')
ax[1].grid();
```
